```python
columns_names = ["Sample code number",
                 "Clump Thickness",
                 "Uniformity of Cell Size",
                 "Uniformity of Cell Shape",
                 "Marginal Adhesion",
                 "Single Epithelial Cell Size",
                 "Bare Nuclei",
                 "Bland Chromatin",
                 "Normal Nucleoli",
                 "Mitoses",
                 "Class"
                ]
datos = pd.read_csv(filepath_or_buffer="breast-cancer-wisconsin.data.txt", names= columns_names)
```

10. reemplace los valores Nan por la media en las columnas numericas y almacenelo en el dataframe original


```python
datos.fillna(datos.mean(), inplace = True)
datos
```

11.  Construya una variable dummy para la columna class (0= benigno, 1= maligno) en un Dataframe llamado DF_class

```python
dummy = ["benigno","maligno"]
DF_class= pd.get_dummies(datos["Class"])
DF_class.columns = dummy
DF_class
```

12. una el DF_class como columnas del Df datos

```python
datos = pd.concat([datos, DF_class], axis = 1)
datos
```
